```python
para='Your Name, directed by Makoto Shinkai, tells the story of two teenagers, Mitsuha and Taki, who mysteriously swap bodies. As they navigate each others lives and form a connection through their shared experiences, they embark on a journey to unravel the supernatural phenomenon binding them together. Set against the backdrop of picturesque landscapes and intertwined destinies, Your Name explores themes of love, fate, and the transcendent power of human connection.'
```

```python
para
```

```python
para.split()
```

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```python
sent=sent_tokenize(para) #splitting the text into sentences.
```

```python
sent[1]
```

```python
word=word_tokenize(para) #text into words.
```

```python
word
```

```python
'''
Stopwords are commonly used words in a language that are typically filtered out 
during text preprocessing because they are considered to have little semantic 
value and do not contribute much to the meaning of a sentence. 
Examples of stopwords in English include "the", "is", "and", "in", "of", etc
'''
from nltk.corpus import stopwords
```

```python
swords=stopwords.words('english')
```

```python
swords
```

```python
x=[wo for wo in word if wo not in swords]
```

```python
x
```

```python
x=[wo for wo in word if wo.lower() not in swords]
```

```python
x
```

```python
#  Stemming is the process of reducing words to their root or base form.
#  PorterStemmer is a stemming algorithm 
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem('working')
```

```python
y=[ps.stem(word) for word in x]
```

```python
y
```

```python
'''
 While stemming involves removing prefixes and suffixes 
 from words to get their root form (stem), lemmatization 
 involves determining the base or dictionary form of a word (lemma)
 based on its intended meaning.
'''
from nltk.stem import WordNetLemmatizer
```

```python
wnl= WordNetLemmatizer()
```

```python
'''
pos='n': Noun
pos='v': Verb
pos='a': Adjective
pos='r': Adverb
'''
wnl.lemmatize('working',pos='v')
```

```python
print(ps.stem('went'))
print(wnl.lemmatize('went',pos='v'))
```

```python
z=[wnl.lemmatize(word,pos='v') for word in x]
```

```python
z
```

```python
import string
string.punctuation 
```

```python
# Using list comprehension to filter out punctuation
t=[wo for wo in word if wo not in string.punctuation] #
```

```python
t
```

```python
'''
takes a list of words as input and assigns a part-of-speech
tag to each word based on its context and usage in a sentence
'''
from nltk import pos_tag
```

```python
pos_tag(t)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
```

```python
v=tfidf.fit_transform(t)
```
