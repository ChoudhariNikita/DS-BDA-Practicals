```python
#Used to identify missing or null values in a DataFrame. 
#where each element is True if the corresponding element in the original DataFrame is null, and False otherwise
df.isnull()
```

```python
# Count number of missing values from each column
df.isnull().sum()
```

```python
#drop all rows that having no value
df.dropna()
```

```python
#Used to replace missing values (NaN) in the DataFrame with a specified value, in this case, 0.
df.fillna(0)
```

```python
#filling the missing values (NaN) in the 'class' column with the string 'TE'.
df['class'].fillna('TE')
```

```python
#filling the missing values (NaN) in the 'marks' column with the mean of the existing values in that column.
df['marks'].fillna(df['marks'].mean())
```

```python
# filling the missing values (NaN) in the 'age' column with the median of the existing values in that column.
df['age'].fillna(df['age'].median())
```

```python
# counts of unique values in the 'class' column of the DataFrame df
df['class'].value_counts()
```

```python
#fill the missing values (NaN) in the 'class' column with the mode (most frequent value) of that column
df['class'].fillna(df['class'].mode()[0])
```

```python
#fill missing values (NaN) in the DataFrame df by propagating non-null values backward along each column.
df.bfill()
```

```python
#fill missing values (NaN) in the DataFrame df by propagating non-null values forward along each column
df.ffill()
```
