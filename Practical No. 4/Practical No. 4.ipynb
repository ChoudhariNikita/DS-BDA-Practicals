```python
#creates a new DataFrame x by dropping the 'medv' column from 
# the original DataFrame df along the specified axis (axis=1 indicates columns)
x=df.drop('MEDV',axis=1) 

y=df['MEDV'] #extracts the 'medv' column from the original DataFrame
```

```python
from sklearn.model_selection import train_test_split

#x and y are your feature and target variables, respectively.
# random_state is used to ensure reproducibility of the split. 
# test_size specifies the proportion of the dataset to include in the test split. 

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0, test_size=0.25)
```

```python
x_train
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# provides simple and efficient tools for data mining and data analysis. 
# It features algos for regression, clustering, dimensionality reduction,
# model selection, and preprocessing,etc.
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression() 
```

```python
# training the regressor model using the provided features (X_train) and their corresponding target values (y_train)
regressor.fit(x_train,y_train)
```

```python
regressor.coef_  #refers to the coefficients of the linear regression model.
```

```python
regressor.intercept_ # refers to the intercept term (B0)
```

```python
# y_pred contains the predicted values of the target variable (y) 
# using the linear regression model regressor on the test dataset x_test.
y_pred=regressor.predict(x_test)
```

```python
y_pred.shape
```

```python
# result is a DataFrame that contains two columns: 'Actual' and 'Predicted'.
result=pd.DataFrame({'Actual':y_test,'Producted':y_pred})
```

```python
result
```

```python
#  calculates the absolute difference between the actual target 
# values (y_test) and the predicted target values (y_pred).
residual_error=abs(y_test-y_pred)
```

```python
residual_error
```

```python
sum(residual_error)/len(residual_error) # calculates the mean of the residual errors
```

```python
# calculates the mean absolute error (MAE) between the predicted values (y_pred) and the actual values (y_test).
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_pred,y_test)
```
